A signal of the form
    
    y(x)=8cos(x)-3sin(7x)
is sent by a kabel and disrupted by noise. We wish to alter the signal, which is recieved so as to minimize the noise.

```python
import numpy as np

def y(x): # y(x) implemented in python
    return 8 * np.cos(x) - 3 * np.sin(7*x)
```

#### (a) Plot the function over the interval [0,9]

```python
import matplotlib.pyplot as plt

xMin = 0
xMax = 9
n   = 100

x = np.linspace(xMin, xMax, n)

fig, ax = plt.subplots()
ax.plot(x, y(x))
```

#### (b) Use the following snippet to add noise and plot the result.

```python
rng = np.random.default_rng()
noise = rng.standard_normal(n)

y_noisy = y(x) + noise

fig, ax = plt.subplots()
ax.plot(x, y_noisy)
```

#### (c) Create a matrix with 1/3's in the diagonals, using diag() and ones()

```python
def thirdsDiag(n, entryVal = 1/3, thickness = 3): # Returns a n*n matrix with 1/3 (or entryVal, if specified) in the diagonal
    # Create 2 vectors with all ones.
    v_1 = np.ones((1,n-1)) # Remove one to fit in the matrix
    v_2 = np.ones((1,n  ))

    # Make the vectors into diagonals
    r_1 = np.diag(v_1[0], 1)
    r_2 = np.diag(v_2[0], 0)
    r_3 = np.diag(v_1[0], -1)
    
    # Add the diagonals together and perform scalar multiplication
    return (r_1+r_2+r_3)*entryVal

print(thirdsDiag(5))
```

### (d) Plot $Ay_{noisy}$. Why is it closer to $y$ than to $y_{noisy}$? 

```python
A = thirdsDiag(n)
Ay_noisy = A@(y(x) + noise)

fig, ax = plt.subplots()
ax.plot(x, Ay_noisy)
```

#### Why is it closer to $y$ than $y_{noise}$?

An admitedly loose explanation would be that, in general a diagonal matrix will preserve the matrix $y_{noise}$, since $AI=A$ where $I$ is the identity matrix. The result of 'thickening' the diagonal, is that we will smoothe out any inconsistencies of the function. We put $\frac{1}{3}$ in the entries, because the 'thickness' of the diagonal is 3 in this case and this preserves the amplitude of the signal.

```python
temp = A

bestSum = 99999
bestPow = 0

for i in range(100):
    currentSum = np.abs( ((temp @ y_noisy)- y(x)).sum() ) 
    if (currentSum < bestSum):
        bestSum = currentSum
        bestPow = i
    temp = A@temp

print(bestSum)
print(bestPow)
```
